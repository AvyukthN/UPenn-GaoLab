```python
from Bio import SeqIO

def load_fasta(fasta_file):
    # Load the entire FASTA file into a dictionary for quick access
    fasta_dict = {}
    for record in SeqIO.parse(fasta_file, "fasta"):
        fasta_dict[record.id] = record.seq
    return fasta_dict

def extract_trimer(seq_dict, location):
    # Parse the location string
    chromosome, pos_range = location.split(':')
    start_pos, end_pos = map(int, pos_range.split('-'))

    if start_pos != end_pos:
        raise ValueError(f"Start and end positions should be the same for location: {location}")

    # Identify the relevant sequence in the dictionary
    for key in seq_dict:
        chrom_key, range_key = key.split(':')
        start_range, end_range = map(int, range_key.split('-'))
        
        # Check if the location falls within the sequence range
        if chrom_key == 'chr' + chromosome and start_range <= start_pos <= end_range:
            # Calculate the index relative to the sequence
            relative_pos = start_pos - start_range
            
            # Extract the trimer centered at the given location
            trimer_start = relative_pos - 1
            trimer_end = relative_pos + 1
            trimer = seq_dict[key][trimer_start:trimer_end+1]
            return str(trimer)
    
    # If no matching sequence is found
    raise ValueError(f"Location {location} not found in the provided sequences.")
    
def get_reverse_complement(seq: str):
    comp_hash = {
        "A": "T",
        "C": "G",
        "G": "C",
        "T": "A"
    }

    new_seq = ""
    for n in seq[::-1]:
        new_seq += comp_hash[n]

    return new_seq
```

```python
### INVESTINGATING SPECIAL CASES OF stop_gained ANNOTATION

def induce_mutation(seq: str, pos: int, mutated_nuc: str):
    seq = list(seq)
    seq[pos] = mutated_nuc
    return "".join(seq)
    
def toRNA(seq: str):
    seq = list(seq)
    for i in range(len(seq)):
        if seq[i] == "T":
            seq[i] = "U"
    return "".join(seq)

def containsStopCodon(seq: str):
    for stop_codon in ["UAA", "UAG", "UGA"]:
        if stop_codon in seq:
            return 1
    return 0


stopg = transcripts[transcripts["Consequence"] == "stop_gained"]
joined_col = [f"{trimer}=>{mut_trimer}" for trimer, mut_trimer in zip(stopg["Trimer"], stopg["MUTATED_TRIMER"])]
stopg.insert(0, "JOINED", joined_col)
stopg_set = stopg.drop_duplicates(subset='JOINED', keep='first')

## EXTRACTING A BUFFER AROUND THE TRIMER FOR ANALYSIS
fasta_dict = load_fasta(f"/Users/avyukthnilajagi/Documents/Learn/UPennGaoLab/data/extracts/2024-08-21-12:04:05.675314_extract/geneFASTAs/{gene}.fa")
# fasta_dict = load_fasta(f"./data/fasta_data/hg38.fa")
buffer = 2
bcol_name = f"BUFFER_TRIMER={buffer}"
buffered_ctrimers = list(map(lambda x: extract_tbuffer(fasta_dict, x, buffer), list(stopg_set["Location"])))
stopg_set.insert(0, bcol_name, buffered_ctrimers)

## Reproducing Mutation in the Buffer Column 
## (buffer column is taken from the original fasta and doesnt contain mutations induced by insilico mutator)
mut_buffers = [induce_mutation(btrimer, buffer, mnuc) for btrimer, mnuc in zip(buffered_ctrimers, stopg_set["ALT"])]
stopg_set.insert(0, f"MUTATED_BUFFER={buffer}", mut_buffers)

## CREATING A COLUMN THAT IS THE REVERSE COMPLEMENT OF THAT BUFFER
rbcol_name = f"REVCOMP_MBUFFER={buffer}"
revcomp_mbuffer = list(map(get_reverse_complement, mut_buffers))
revcomp_mbuffer = list(map(toRNA, revcomp_mbuffer))
stopg_set.insert(0, rbcol_name, revcomp_mbuffer)

## Annotating which buffers contain stop codons
containsStop = list(map(containsStopCodon, revcomp_mbuffer))
stopg_set.insert(0, "containsStop", containsStop)

## Filter out 
stopg_set = stopg_set[stopg_set["containsStop"] == 0]

# values_to_exclude = ['=>None']
# stopg_set = stopg_set[~stopg_set['JOINED'].isin(values_to_exclude)]
stopg_set.to_csv("./BRCA2-stop_gained-special_cases.tsv", sep="\t")
stopg_set
```
